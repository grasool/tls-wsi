```python
import matplotlib.pyplot as plt
```

```python
from PIL import Image
import numpy as np

# Load the images
jpg_1 = Image.open(r'D:\Data\tls-data\patched-data1\train_gt\C3L-00415-2183_102.jpg')
tif_1 = Image.open(r'D:\Data\tls-data\patched-data\train_gt\C3L-00415-2183_102.tif')

# Convert the images to numpy arrays
img1_array = np.array(jpg_1)
img2_array = np.array(tif_1)

# Compare the arrays
if np.array_equal(img1_array, img2_array):
    print("The images have the same contents.")
else:
    print("The images have different contents.")
```

```python
# Normalize the images to [0, 1]
img1_array = img1_array / 255.0
img2_array = img2_array / 255.0

# Display the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img1_array)
plt.title('Image 1')

plt.subplot(1, 2, 2)
plt.imshow(img2_array)
plt.title('Image 2')

plt.show()
```
